```python
# Grade Tracker and Character Counter Program
# This program allows users to manage grades and count characters in a string

#1. Naming Conventions/Program Readability: lines 17-19, 59, 61, 63,   ex. line 18 set_password = "password123"
#2. Basic Data Types and Operators: lines 49, 50, 61, 63, 65,          ex. line 61 print("Your average grade is:", float(average))
#3. Input: lines 19, 22, 28, 34, 44,                                   ex. line 19 user_password = input("Enter the password: ")
#4. Output: lines 21, 23, 37, 46, 54, 56, 61, 63, 65, 67, 73-74, 77,   ex. line 23 print("The password is correct. You may proceed.")
#5. Error Handling: lines 36, 51, 53, 66,                              ex. line 36-37 else: print("Enter 'gc', 'cc', or 'q'")
#6. Password: lines 17-23,                                             ex. line 18 set_password = "password123"
#7. Functions: lines 13, 26, 40, 70,                                   ex. line 26 def options():
#8. Conditional Statements: lines 29, 45, 49, 57,                      ex. line 49-52 if int(user_input) <= 100 and int(user_input) >= 0: else: print("Enter a grade within 0-100")
#9. Loops: lines 20, 27, 43,                                           ex. line 20 while user_password != set_password:
#10. Advanced Data Types: line 72,                                     ex. line 72 input_list = list(user_input)
#11. Search: line 58, 62,                                              ex. line 58 for grade in grades:

# This function allows you to continue when the password input matches the set password
def password():
    set_password = "password123"
    user_password = input("Enter the password: ")
    while user_password != set_password:
        print("The password is incorrect. Try again.")
        user_password = input("Enter the password: ")
    print("The password is correct. You may proceed.")

# This function allows you to choose between a grade calculator program, a character counting program, or quitting
def options():
    while True:
        user = input("Enter 'gc' for grade calculator, 'cc' for character count, or 'q' for quit: ")
        if user == "q":
            break
        elif user == "gc":
            grade_calc()
        elif user == "cc":
            user = input()
            count_char(user)
        else:
            print("Enter 'gc', 'cc', or 'q'")

# This function calculates the average of grades, counts hundreds, and prints sorted grades if user enters a value between 0 and 100
def grade_calc():
    grades = []
    sum_grades = 0
    while True:
        user_input = input("Enter a grade or 'q' to quit ")
        if user_input.startswith("q"):
            print("Quitting grade calculator")
            break
        elif user_input.isdigit():
            if int(user_input) <= 100 and int(user_input) >= 0:
              grades.append(int(user_input))
            else:
              print("Enter a grade within 0-100")
        else:
            print("Please enter a numerical value")
    grades.sort()
    print("Here is a sorted list of your grades: ", grades)
    if len(grades) > 0:
        for grade in grades:
            sum_grades += grade
        average = sum_grades / len(grades)
        print("Your average grade is:", float(average))
        hundreds_count = grades.count(100)
        print("You have " + str(hundreds_count) + " hundred(s)")
        sum_hundreds = 100 * hundreds_count
        print("Total points from hundreds: " + str(sum_hundreds))
    else:
        print("No grades entered")

# This function counts and displays the number of characters in the input string from a parameter
def count_char(user_input):
    input_length = len(user_input)
    input_list = list(user_input)
    print(input_length, "characters in", user_input)
    print(input_list)

# Main program that runs after password is correct
print("The password is: password123")
password()
options()
```
